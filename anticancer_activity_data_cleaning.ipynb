```python
# import libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# target protein
target = new_client.target
target_query = target.search('aromatase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.columns
```

```python
df.standard_type.unique()
```

```python
df.dropna(subset=['standard_value'], inplace=True)
```

```python
df.standard_units.unique()
```

```python
unit_counts = df['standard_units'].value_counts()
print("Counts of standard_units:")
print(unit_counts[['nM', 'mg kg-1', 'ug.mL-1']])
```

```python
df = df[~df['standard_units'].isin(['mg kg-1', 'ug.mL-1'])]
display(df)
```

```python
df.standard_units.unique()
```

```python
df['standard_value'] = pd.to_numeric(df['standard_value'], errors='coerce') # All the IC50 Results as Numbers
#Classify them into Active, Intermediate, Inactive
# (0-10) Active, (10-30) Intermediate, (30-1000) Inactive
# rigt = False means higher than 0 and lower than 10, etc
df['activity'] = pd.cut(df.standard_value, bins=[0, 1000, 10000, 1000000], labels=['active', 'intermediate', 'inactive'], right= False)
```

```python
# Create another Data Frame with only 'molecule_chembl_id', 'canonical_smiles', 'standard_value' Columns
df2 = df[['canonical_smiles', 'standard_value', 'standard_value', 'activity']].copy()

#Rename Columns
df2.rename(columns={'canonical_smiles':'SMILES', 'standard_value': 'IC50 nM', 'molecule_chembl_id': 'ChEMBL'},inplace=True)

df2.head() # Show the Head of the File
```

```python
df2.to_csv('bioactivity_data.csv', index=False)
```
